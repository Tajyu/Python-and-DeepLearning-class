```python
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)  # warning 출력 방지
from keras.applications.resnet50 import ResNet50, decode_predictions
resnet = ResNet50()
resnet.summary()
```

```python
import cv2
import timeit
# import time 

img = cv2.imread('bird.jpg', -1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
img = cv2.resize(img, (224, 224))
```

```python
start = timeit.default_timer()
print("Program start ")
# ALL THE PROGRAM STATEMETNS
# run the experiment - pool_size
yhat = resnet.predict(img.reshape(-1, 224, 224, 3))

stop = timeit.default_timer()
execution_time = stop - start

# print("Program Executed in " + str(execution_time))

label = decode_predictions(yhat)
label = label[0][0]

print("테스트 시 소요 시간 : {}".format(execution_time))
print('%s (%.2f%%)' % (label[1], label[2]*100))
img = img[:,:,::-1]
plt.figure(figsize=(11,11))
plt.imshow(img)
plt.axis("off")
plt.show()
```
